```python
df_1 = pd.read_csv('https://raw.githubusercontent.com/PabloJMoreno/dataset/main/MSFT.csv', header = 0)
df_1
df_1.info()
```

```python
df_1['Date'] = pd.to_datetime(df_1['Date'])
df_1.info()
```

```python
# let's create a new dataframe object as we are going to make changes, so we always preserve the original dataset
df_time = df_1[['Date', 'Close']]
df_time.head()
df_time.info()
```

```python
# all those transformation are possible because column 'Date' is now as date format
# we will create multiple new columns with the components of each date per row

df_time['only_date'] = df_time['Date'].dt.date
df_time['time'] = df_time['Date'].dt.time
df_time['y'] = df_time['Date'].dt.year
df_time['q'] = df_time['Date'].dt.quarter
df_time['m'] = df_time['Date'].dt.month
df_time['d'] = df_time['Date'].dt.day
df_time['h'] = df_time['Date'].dt.hour
df_time['min'] = df_time['Date'].dt.minute
df_time['sec'] = df_time['Date'].dt.second
df_time['woy'] = df_time['Date'].dt.weekofyear
df_time['dow'] = df_time['Date'].dt.weekday
df_time['doy'] = df_time['Date'].dt.dayofyear
df_time['fdom'] = df_time['Date'].dt.is_month_start
df_time['ldom'] = df_time['Date'].dt.is_month_end
df_time['fdoq'] = df_time['Date'].dt.is_quarter_start
df_time['ldoq'] = df_time['Date'].dt.is_quarter_end
df_time['fdoy'] = df_time['Date'].dt.is_year_start
df_time['ldoy'] = df_time['Date'].dt.is_year_end
df_time['d_m'] = df_time['Date'].dt.daysinmonth
df_time['d_n'] = df_time['Date'].dt.day_name()
df_time['m_n'] = df_time['Date'].dt.month_name()
```

```python
# Let's review the result of the transformed dataframe
df_time
```

```python
df_left = pd.DataFrame(
    {"cliente": ["Carlos", "Luis", "Maria", "Susana", "Camila"],
     "region": ["Norte", "Norte", "Sur", "Sur", "Sur"]})
df_left
df_right = pd.DataFrame(
    {"cliente": ["Carlos", "Luis", "Maria", "Susana", "Pedro"],
     "compras": [1500, 300, 3000, 2000, 500]})
df_right
```

```python
pd.merge(df_left, df_right, how = "left", on = "cliente")
```

```python
pd.merge(df_left, df_right, how = "right", on = "cliente")
```

```python
pd.merge(df_left, df_right, how = "outer", on = "cliente")
```

```python
pd.merge(df_left, df_right, how = "inner", on = "cliente")
```
